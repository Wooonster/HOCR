```python
import cv2
import io
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### 模拟

将图片转为 byte 格式 --> 转回 image --> grayscale --> 二值化

```python
def simulate(img):
    ' 转为 PIL 格式 '
    pil_img = Image.fromarray(img)

    ' BytesIO '
    img_stream = io.BytesIO()
    pil_img.save(img_stream, format='PNG')

    img_bytes = img_stream.getvalue()

    ' io 读取 bytes'
    read_img_bytes = io.BytesIO(img_bytes)
    ' 读取 灰度图 '
    read_img = Image.open(read_img_bytes).convert('L')
    # read_img.show()
    read_img_arr = np.array(read_img)
    ' 二值化 '
    _, img_bin = cv2.threshold(read_img_arr, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    img_thresholded = Image.fromarray(img_bin)
    return img_bytes, img_thresholded
```

```python
' 读取图片 '

img_paths = [
    'test/imgs/eq1.png',
    'test/imgs/hard_1.png',
    'test/97628.jpg',
    'test/train_281.jpg',
    'test/train_1009.jpg',
    'test/train_1845.jpg',
]


imgs = [cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB) for path in img_paths]

plt.figure()
for i in imgs:    
    plt.imshow(i)
    plt.show()
```

```python
plt.figure()
' 模拟 '
for i in imgs:
    bytes, thresholded = simulate(i)
    print(bytes)
    plt.imshow(thresholded, cmap='gray')
    plt.show()
```
